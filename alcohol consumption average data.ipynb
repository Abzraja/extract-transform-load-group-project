```python
import pandas as pd
import re
```

```python
alcohol_dataset=pd.read_csv("Resources/WHO-alcohol-data.csv")
alcohol_dataset.head()
```

```python
alcohol_dataset.columns
```

```python
dropped_df=alcohol_dataset.drop(columns=["IndicatorCode","Indicator","ValueType","ParentLocationCode", "ParentLocation", "ParentLocation","Location type", "SpatialDimValueCode","Period type", "IsLatestYear", "Dim1", "Dim3", "Dim1 type", "Dim1ValueCode", "Dim2 type", "Dim2", "Dim2ValueCode", "Dim3 type", "Dim3 type", "Dim3ValueCode", "DataSourceDimValueCode", "DataSource",
       "FactValueNumericPrefix", "FactValueNumeric", "FactValueUoM",
       "FactValueNumericLowPrefix", "FactValueNumericLow",
       "FactValueNumericHighPrefix", "FactValueNumericHigh",
       "FactValueTranslationID", "FactComments", "Language", "DateModified" ])
dropped_df
```

```python
renamed_df = (dropped_df.rename(columns={"Location":"country", "Period":"year", "Value":"litres per capita"}))
renamed_df.head()
```

```python
# only want 2015 and 2019
condition = (renamed_df["year"]==2015) | (renamed_df["year"]==2019)
df_filtered = renamed_df.loc[condition, :]
```

```python
df_filtered['litres per capita average'] = df_filtered["litres per capita"].apply(lambda x: x.split("[")[0].strip())
```

```python
switzerland = df_filtered.loc[df_filtered["country"] == "Switzerland"]
switzerland
```

```python
df_filtered['litres per capita average'] = pd.to_numeric(df_filtered['litres per capita average'])
df_filtered.dtypes
```

```python
df_filtered = df_filtered.drop(columns=["litres per capita"])
```
