```python
# import pandas
import pandas as pd
```

```python
# load our cleaned and transformed happiness data
happiness_data =  pd.read_csv("Resources/Output/Happiness_Data_Transformed.csv")
```

```python
# display the dataframe
happiness_data
```

```python
# drop the Unnamed:0 column that has appeared.
happiness_data = happiness_data.drop(columns=["Unnamed: 0"])
```

```python
# display the dataframe
happiness_data
```

```python
# load our cleaned and transformed alcohol consumption data
alcohol_data = pd.read_csv("Resources/Output/Alcohol_Data_Transformed.csv")
```

```python
# display the data frame
alcohol_data
```

```python
# drop the new Unnamed:0 column that has appeared
alcohol_data = alcohol_data.drop(columns=["Unnamed: 0"])
```

```python
# set the country as the index
alcohol_data = alcohol_data.set_index("country")
```

```python
# display the data frame
alcohol_data
```

```python
# merge the 2 imported Alcohol Consumption Data and Happiness Data on Country and Year
merged_df = happiness_data.merge(alcohol_data, on=["country","year"])
```

```python
#display the data frame
merged_df
```

```python
#reorder columns
merged_df = merged_df[["country","year","litres per capita average","happiness score","gdp per capita","healthy life expectancy"]]
```

```python
#check for null values
merged_df.isnull().sum()
```

```python
# check values
merged_df.value_counts()
```

```python
# set country as the index
merged_df = merged_df.set_index("country")
merged_df
```

```python
# rename columns for loading into Postgre SQL - I think spaces are not supported so have replaced with underscores.
merged_df = merged_df.rename(columns={"litres per capita average":"litres_per_capita_average", "happiness score":"happiness_score", "gdp per capita":"gdp_per_capita", "healthy life expectancy":"healthy_life_expectancy"})
merged_df
```

```python
#sort by litres alcohol consumed descending
merged_df = merged_df.sort_values(by=["litres_per_capita_average"], ascending=False)
merged_df
```

```python
# export to CSV
merged_df.to_csv("Resources/Final_Merged_Data.csv")
```
