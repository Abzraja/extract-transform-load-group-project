```python
import pandas as pd
```

```python
Happiness_2015_df = pd.read_csv("Resources/2015.csv")
Happiness_2015_df
```

```python
transformed_Happiness_2015_df=Happiness_2015_df.drop(columns=['Happiness Rank','Region','Standard Error','Family','Dystopia Residual','Freedom','Trust (Government Corruption)','Dystopia Residual','Generosity'])
renamed_Happiness_2015_df=transformed_Happiness_2015_df.rename(columns={'Country':'country',
 'Happiness Score':'happiness score','Economy (GDP per Capita)':'gdp per capita',
'Health (Life Expectancy)': 'healthy life expectancy'})
renamed_Happiness_2015_df['year']=2015
renamed_Happiness_2015_df
```

```python
Happiness_2019_df = pd.read_csv("Resources/2019.csv")
Happiness_2019_df
```

```python
transformed_Happiness_2019_df=Happiness_2019_df.drop(columns=['Overall rank','Freedom to make life choices', 'Generosity','Perceptions of corruption','Social support'])
renamed_Happiness_2019_df=transformed_Happiness_2019_df.rename(columns={'Country or region' : 'country', 'Score' : 'happiness score', 'GDP per capita' : 'gdp per capita',
'Healthy life expectancy': 'healthy life expectancy'})
renamed_Happiness_2019_df['year']=2019
renamed_Happiness_2019_df
```

```python
merged_df=pd.concat([renamed_Happiness_2015_df, renamed_Happiness_2019_df])
merged_df
```

```python
merged_df.to_csv("Resources/Happiness_Data_Transformed.csv")
```
